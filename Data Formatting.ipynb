## Clean Up Data 1

```python

import pandas as pd

path = r"C:\Users\BGROSS\Downloads\essex.xlsx"
df = pd.read_excel(path,sheet_name='ESSEX COUNTY 2022' , header=1)
df = df.dropna(axis=1, how='all')

df = df[~df['o'].isin([0.1,0.2])]


df.head(10)
```

```python
df.columns = df.columns.str.replace('Unnamed: 7', 'MUNI')
df.head()
```

```python
df['MUNI'] = df['MUNI'].fillna(method='ffill')
df = df.dropna(axis=0,subset=['seq'])
df.head(10)
```

## Format Data Set 2

```python
muni_codes = pd.read_excel(r"C:\Users\BGROSS\Downloads\simplified data 7.xlsx")
muni_codes = muni_codes[['Muni Code','Municipality']]
muni_codes['Municipality'] = muni_codes['Municipality'].str.split(' ')
muni_codes['Municipality'] = muni_codes['Municipality'].apply(lambda x: " ".join(x[:-1]).strip() )
muni_codes['Muni Code'] = muni_codes['Muni Code'].astype(int)
muni_codes.head(20)
```

```python
# df['MUNI'] = df['MUNI'].astype(int)
df.join( muni_codes.set_index("Municipality"), on='MUNI', how='left' )

df.to_csv('output.csv')

```
